```python
df['Income'] = df['Income'].fillna(df['Income'].mean())
```

```python
sns.heatmap(df[['Income','Total_expenses','Age','Total_acc_cmp','Recency']].corr(),annot=True)
```

```python
from sklearn.preprocessing import LabelEncoder
lenc = LabelEncoder()
lenc.fit(df['Marital_Status'])
df['Marital_Status'] = lenc.transform(df['Marital_Status'])
from sklearn.preprocessing import OrdinalEncoder
edu = ['Basic', 'Graduation', 'Master', '2n Cycle', 'PhD']
ore = OrdinalEncoder(categories=([edu]))
ore.fit(df[['Education']])
df['Education'] = ore.transform(df[['Education']])
```

```python
df_prep = df.drop(['ID', 'Year_Birth','Dt_Customer','AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2','Response', 'Complain', 'Z_CostContact', 'Z_Revenue'],axis=1)
```

```python
df_prep.columns
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
s_d = np.array(df_prep[['Income', 'Kidhome', 'Teenhome',
       'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts',
       'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
       'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
       'NumStorePurchases', 'NumWebVisitsMonth', 'Age', 'Total_expenses',
       'Total_acc_cmp', 'Years_customer']]).reshape(-1,19)
sc.fit(s_d)
df_prep[['Income', 'Kidhome', 'Teenhome',
       'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts',
       'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
       'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
       'NumStorePurchases', 'NumWebVisitsMonth', 'Age', 'Total_expenses',
       'Total_acc_cmp', 'Years_customer']] = sc.transform(s_d)
```

```python
df_prep
```

```python
from sklearn.cluster import KMeans
l=[]
for i in range(1,7):
    kmeans = KMeans(i)
    kmeans.fit(df_prep)
    inertia = kmeans.inertia_
    l.append(inertia)
    
number_clusters = [1,2,3,4,5,6]
plt.figure(figsize=(10,6))
plt.plot(number_clusters, l, '-o')
plt.title('Elbow Plot')
plt.xlabel('Number Of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
model = KMeans(2)
model.fit(df_prep)
cluster = model.fit_predict(df_prep) 
df_segment = df_prep.copy()
df_segment['Labels'] = cluster
plt.figure(figsize=(10,8))
sns.scatterplot(x='Income',y='Total_expenses',data=df_segment,hue='Labels',palette=['g','r'])
plt.title("Segmentation KMeans")
plt.show()
```
